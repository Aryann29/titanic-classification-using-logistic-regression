```python
old_df = pd.read_csv('/content/drive/MyDrive/Datasets/titanic/train.csv')
```

```python
df = old_df
```

```python

df = df.drop(['PassengerId',"Cabin"],axis=1)
```

```python
df.isna().sum()
```

```python
df[df.Name.str.contains('Master') & df.Age.isnull()]
```

```python
import math
c_agemean = math.floor(df[df.Name.str.contains('Master')] ['Age'].mean())
male_agemean = math.floor(df[df.Name.str.contains('Mr')] ['Age'].mean())
female_agemean = math.floor(df[df.Name.str.contains('Mrs')] ['Age'].mean())
```

```python
girls = math.floor(df[df.Name.str.contains('Miss')] ['Age'].mean())
```

```python
df[df.Name.str.contains('Master') & df.Age.isnull()] = df[df.Name.str.contains('Master') & df.Age.isnull()].fillna(c_agemean)
```

```python
df[df.Name.str.contains('Mr') & df.Age.isnull()] = df[df.Name.str.contains('Mr') & df.Age.isnull()].fillna(male_agemean)
```

```python
df[df.Name.str.contains('Miss') & df.Age.isnull()] = df[df.Name.str.contains('Miss') & df.Age.isnull()].fillna(girls)
```

```python
df.isna().sum()
```

```python
df[df.Name.str.contains('Dr') & df.Age.isnull()] = df[df.Name.str.contains('Dr') & df.Age.isnull()].fillna(33)
```

```python
df.isna().sum()
```

```python
df["Embarked"] = df['Embarked'].fillna("S")
```

```python
df.isna().sum()
```

```python
df.Embarked.mode()
```

```python
df
```

```python
df.replace({"Sex" :{'male':0 ,"female":1},"Embarked" :{'S':0 ,"Q":1 , "C":2}},inplace=True)
```

```python
df
```

```python
X = df[['Pclass','Sex',"SibSp","Parch",'Fare','Age',"Embarked"]]
X
```

```python
Y = df[['Survived']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
lr =LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
# accuracy on training data
X_train_prediction = lr.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
X_test_prediction = lr.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
